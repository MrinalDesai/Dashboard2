```python
import pandas as pd
```

```python
df=pd.read_csv("stop_data.csv")
```

```python
df
```

```python
v = df.values.tolist()
c = df.columns.values.tolist()

final=[dict(zip(c, x)) for x in v]
```

```python
type(final)
```

```python
final[0]
```

```python
Dict = { }
New_Dict=[]
for i in final:
  Dict = { }
  Dict['data'] = i
  New_Dict.append(Dict)
```
